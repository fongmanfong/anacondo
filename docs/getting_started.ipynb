### Run Scenario Analysis

It's important to understand investment return under different assumption scenarios because it provides you with a sense so sensitivity and risk and also your "worst" case scenario

```python
yorkville_scenarios = ScenarioAnalyzer(yorkville)
```

```python
# Another important view is that the parameters we've defined are estimates. Meaning there is a level of uncertainty
# Because humans are exceptionally bad at point estimates. Instead, we should estimate a range with some level of uncertainty
# E.g. I did research that rent in this area is between X and Y with Z being most observed.
# Here we pass in a distribution for parameters we are uncertain and run a montecarlo simulation to get estimates

import numpy as np
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

size = 1000
param_grid = {'interest_rate':  np.random.normal(loc=0.035, scale=0.005, size=size),
              'purchase_price': np.random.normal(loc=620000, scale=10000, size=size),      
              'rental_income': np.random.normal(loc=2500, scale=100, size=size),
              'monthly_property_tax': np.random.normal(loc=150, scale=15, size=size),
              'monthly_insurance': np.random.normal(loc=60, scale=10, size=size)
             }
```

```python
# Understand the return after 2 and 5 years for cash on cash return

five_year_return = yorkville_scenarios.compute_scenarios_mc('coc', param_grid, year=5, sample_size=size)
two_year_return = yorkville_scenarios.compute_scenarios_mc('coc', param_grid, year=2, sample_size=size)
```

```python
figsize(12, 5)
plt.hist(five_year_return, alpha=0.8, label='5 Year');
plt.hist(two_year_return, alpha=0.8, label='2 Year');
plt.legend()
```

```python
print ('Probabilty of positive return after 2 years: {0}%'.format(100.0*sum(two_year_return > 0)/len(two_year_return)))
print ('Probabilty of positive return after 5 years: {0}%'.format(100.0*sum(five_year_return > 0)/len(five_year_return)))
```
